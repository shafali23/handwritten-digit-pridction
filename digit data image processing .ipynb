```python
x,y = digit.data[:-1], digit.target[:-1]
```

```python
clf = svm.SVC(gamma = 0.001, C=100)
#SVc for classification, svr for regression
#gamma is for precision of classification
#C and Gamma are the parameters for a nonlinear support vector machine (SVM) with a Gaussian radial basis function kernel.
#A standard SVM seeks to find a margin that separates all positive and negative examples. gamma value shoul be low.
```

```python
clf.fit(x,y)
```

```python
#plotting the real image and comparing with prediction
plt.imshow(digit.images[-1],cmap = plt.cm.gray_r, interpolation = "nearest")
```

```python
#predicting the image
print("prediction :" ,clf.predict(digit.data[[-1]]))
```
